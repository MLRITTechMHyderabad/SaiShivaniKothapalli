```python
import os
import pandas as pd
import mysql.connector
from mysql.connector import Error
from tabulate import tabulate
import matplotlib.pyplot as plt
from dotenv import load_dotenv
from sqlalchemy import create_engine

# Step 1: Dataset Acquisition
file_path = r'C:\Users\Lenovo\Desktop\revature_training\Projects\Project\cybersecurity-threat-analysis\cybersecurity_threats.csv'

# Step 2: Python File Handling
def load_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print("Data loaded successfully!")
        return df
    except FileNotFoundError:
        print(" File not found. Check the path.")
    except Exception as e:
        print("Error loading data:", e)
    return None

# Step 3: Data Cleaning
def clean_data(df):
    df.columns = (
        df.columns.str.strip()
                  .str.lower()
                  .str.replace(' ', '_')
                  .str.replace('(', '')
                  .str.replace(')', '')
                  .str.replace('$', 'dollar')
    )
    print("Data cleaned!")
    return df

# Step 4: MySQL Integration
def connect_to_db():
    try:
        load_dotenv()
        config = {
            'host': os.getenv('DB_HOST'),
            'user': os.getenv('DB_USER'),
            'password': os.getenv('DB_PASSWORD'),
            'database': os.getenv('DB_NAME')
        }
        conn = mysql.connector.connect(**config)
        if conn.is_connected():
            print("Connected to MySQL...")
        return conn
    except Error as e:
        print("Connection Error:", e)
    return None

def create_table(cursor):
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS threats (
            id INT AUTO_INCREMENT PRIMARY KEY,
            country VARCHAR(100),
            year INT,
            attack_type VARCHAR(100),
            target_industry VARCHAR(100),
            financial_loss FLOAT,
            affected_users BIGINT,
            attack_source VARCHAR(100),
            vulnerability_type VARCHAR(100),
            defense_mechanism VARCHAR(100),
            resolution_time FLOAT
        )
    """)


def insert_data(cursor, df):
    insert_query = """
        INSERT INTO threats (
            country, year, attack_type, target_industry, financial_loss,
            affected_users, attack_source, vulnerability_type,
            defense_mechanism, resolution_time
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    values = [
        (
            row['country'], row['year'], row['attack_type'], row['target_industry'],
            row['financial_loss_in_million_dollar'], row['number_of_affected_users'],
            row['attack_source'], row['security_vulnerability_type'],
            row['defense_mechanism_used'], row['incident_resolution_time_in_hours']
        )
        for _, row in df.iterrows()
    ]
    cursor.executemany(insert_query, values)
    print("Data insertion completed!")

# Step 5: Data Analysis
def analyze_data(df):
    print("\nPerforming data analysis...")

    # Top 10 Countries
    top_countries = df['country'].value_counts().head(10).reset_index()
    top_countries.columns = ['Country', 'Incidents']
    print("\nTop 10 Countries")
    print(tabulate(top_countries, headers='keys', tablefmt='fancy_grid', numalign="right"))

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.bar(top_countries['Country'], top_countries['Incidents'], color='skyblue')
    plt.title("Top 10 Countries with Cybersecurity Threats")
    plt.xlabel("Country")
    plt.ylabel("Incidents")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Top 5 Attack Types
    top_attacks = df['attack_type'].value_counts().head(5).reset_index()
    top_attacks.columns = ['Attack Type', 'Frequency']
    print("\nTop 5 Attack Types")
    print(tabulate(top_attacks, headers='keys', tablefmt='fancy_grid', numalign="right"))

    # Yearly Trends
    yearly = df.groupby('year').size().reset_index(name="Incidents")
    print("\nYearly Trends")
    print(tabulate(yearly, headers='keys', tablefmt='fancy_grid', numalign="right"))

    # Financial Loss by Year
    loss_year = df.groupby('year')['financial_loss_in_million_dollar'].sum().reset_index(name="Loss ($M)")
    print("\nFinancial Loss by Year")
    print(tabulate(loss_year, headers='keys', tablefmt='fancy_grid', numalign="right"))

    plt.figure(figsize=(10, 5))
    plt.plot(loss_year['year'], loss_year['Loss ($M)'], marker='o', color='red')
    plt.title("Financial Loss Over Years")
    plt.xlabel("Year")
    plt.ylabel("Loss in Million $")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Resolution Time by Defense Mechanism
    res_time = df.groupby('defense_mechanism_used')['incident_resolution_time_in_hours'].mean().reset_index(name="Avg Resolution Time (hrs)")
    print("\nResolution Time by Defense Mechanism")
    print(tabulate(res_time, headers='keys', tablefmt='fancy_grid', numalign="right"))

    plt.figure(figsize=(10, 6))
    plt.barh(res_time['defense_mechanism_used'], res_time['Avg Resolution Time (hrs)'], color='green')
    plt.title("Avg Resolution Time by Defense Mechanism")
    plt.xlabel("Time (hrs)")
    plt.ylabel("Defense Mechanism")
    plt.tight_layout()
    plt.show()

    # Correlation
    correlation = df['number_of_affected_users'].corr(df['financial_loss_in_million_dollar'])
    print(f"\nCorrelation between Affected Users and Financial Loss: {correlation:.2f}")
 
   # Severity levels and their impact by region (total financial loss and affected users by country)
    severity_by_region = df.groupby('country')[['financial_loss_in_million_dollar', 'number_of_affected_users']].sum().reset_index()
    severity_by_region = severity_by_region.sort_values(by='financial_loss_in_million_dollar', ascending=False).head(10)
    print("\nSeverity Levels and Their Impact by Region")
    print(tabulate(severity_by_region, headers='keys', tablefmt='fancy_grid', numalign="right"))

    plt.figure(figsize=(10, 6))
    plt.barh(severity_by_region['country'], severity_by_region['financial_loss_in_million_dollar'], color='purple')
    plt.title("Top 10 Countries by Severity (Financial Loss)")
    plt.xlabel("Financial Loss (Million $)")
    plt.ylabel("Country")
    plt.tight_layout()
    plt.show()

    # Correlation between attack type and sector targeted (attack types vs. industries)
    attack_sector_correlation = df.groupby(['attack_type', 'target_industry']).size().reset_index(name="Frequency")
    attack_sector_correlation = attack_sector_correlation.sort_values(by='Frequency', ascending=False).head(10)
    print("\nCorrelation between Attack Type and Sector Targeted")
    print(tabulate(attack_sector_correlation, headers='keys', tablefmt='fancy_grid', numalign="right"))




# Main
def main():
    df = load_data(file_path)
    if df is not None:
        df = clean_data(df)
        analyze_data(df)
        conn = connect_to_db()
        if conn:
            cursor = conn.cursor()
            create_table(cursor)
            insert_data(cursor, df)
            conn.commit()
            cursor.close()
            conn.close()
            print("Data successfully inserted into MySQL!")

if __name__ == "__main__":
    main()
```
